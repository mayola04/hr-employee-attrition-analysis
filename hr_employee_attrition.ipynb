```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
import zipfile
# Set style for plots
sns.set(style="whitegrid")
```

```python
import os
from google.colab import drive
```

```python
# Extract Dataset
zip_path = "/content/drive/MyDrive/archive (1).zip"
extract_to = "/content/dataset"
with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(extract_to)
```

```python
csv_file = os.path.join(extract_to, 'WA_Fn-UseC_-HR-Employee-Attrition.csv')
df = pd.read_csv(csv_file)
```

```python
# Get basic info
print(df.shape)
print(df.columns)

# Summary statistics
df.describe()

# Check for missing values
df.isnull().sum()
```

```python
# Step 5: Explore basic info
print("✅ Dataset Loaded")
print("Rows, Columns:", df.shape)
print("\nMissing Values:\n", df.isnull().sum())
print("\nColumns:\n", df.columns.tolist())
```

```python
# Step 6: Attrition Count
plt.figure(figsize=(6, 4))
sns.countplot(x='Attrition', data=df)
plt.title('Attrition Count')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

```python
# Step 7: Attrition by Department
plt.figure(figsize=(8, 4))
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title('Attrition by Department')
plt.xticks(rotation=30)
plt.xlabel('Department')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.show()
```

```python
# Step 8: Age Distribution by Attrition
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Age', hue='Attrition', multiple='stack', kde=True)
plt.title('Age Distribution by Attrition')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Step 9: Monthly Income vs Attrition
plt.figure(figsize=(8, 5))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title('Monthly Income vs Attrition')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.show()
```

```python
# Step 10: Years at Company Distribution
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df[df['Attrition'] == 'Yes']['YearsAtCompany'], label='Left', shade=True)
sns.kdeplot(data=df[df['Attrition'] == 'No']['YearsAtCompany'], label='Stayed', shade=True)
plt.title('Years at Company - Attrition vs No Attrition')
plt.xlabel('Years at Company')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Step 11: Summary Insights
print("\n Summary Insights:")
print("1. Highest attrition is observed in the Sales department.")
print("2. Employees aged 25–35 are more likely to leave.")
print("3. Employees with lower monthly income tend to leave more.")
print("4. Most employees who leave have spent fewer than 3 years at the company.")
```
